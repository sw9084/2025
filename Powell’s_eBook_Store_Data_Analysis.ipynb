```python
import pandas as pd

# Load dataset
url = "https://docs.google.com/spreadsheets/d/1XSBE7VwTJ-xB300r2JAwZesaAbGTLvMgS4qRxXfDjys/pub?gid=0&single=true&output=csv"
df = pd.read_csv(url)

# Drop unnecessary columns
df_cleaned = df.drop(['Customer ID', 'First Name', 'Middle Name', 'Last Name', 'Suffix',
                      'Street Address1', 'Street Address2'], axis=1)

# Check the cleaned data
df_cleaned.head()
```

```python
# Check for missing values in the cleaned data
missing_values = df_cleaned.isnull().sum()

# Display missing values per column
print(missing_values)
```

```python
# Split 'City-ZipCode-State' into separate columns
df_cleaned[['City', 'ZipCode', 'State']] = df_cleaned['City-ZipCode-State'].str.split('-', expand=True)

# Drop the original combined column
df_cleaned.drop(columns=['City-ZipCode-State'], inplace=True)
```

```python
# Handle missing values by filling them with mean (for numerical) and mode (for categorical)
# For numerical columns, use the mean
df_cleaned['Annual Income'].fillna(df_cleaned['Annual Income'].mean(), inplace=True)
df_cleaned['Avg Monthly Spend'].fillna(df_cleaned['Avg Monthly Spend'].mean(), inplace=True)

# For categorical columns, use the mode
df_cleaned['Education Level'].fillna(df_cleaned['Education Level'].mode()[0], inplace=True)
df_cleaned['Occupation'].fillna(df_cleaned['Occupation'].mode()[0], inplace=True)
df_cleaned['Gender'].fillna(df_cleaned['Gender'].mode()[0], inplace=True)
df_cleaned['Marital Status'].fillna(df_cleaned['Marital Status'].mode()[0], inplace=True)
df_cleaned['City'].fillna(df_cleaned['City'].mode()[0], inplace=True)
df_cleaned['ZipCode'].fillna(df_cleaned['ZipCode'].mode()[0], inplace=True)
df_cleaned['State'].fillna(df_cleaned['State'].mode()[0], inplace=True)

# Verify if there are any missing values left
missing_values_after = df_cleaned.isnull().sum()
print(missing_values_after)
```

```python
import pandas as pd
from datetime import datetime

# Convert the 'Birth Date' column to datetime format
df_cleaned['Birth Date'] = pd.to_datetime(df_cleaned['Birth Date'], errors='coerce')

# Check for invalid (future) birth dates
invalid_birth_dates = df_cleaned[df_cleaned['Birth Date'] > datetime.now()]
print("Invalid birth dates found:\n", invalid_birth_dates)

# Drop rows with invalid birth dates
df_cleaned = df_cleaned[df_cleaned['Birth Date'] <= datetime.now()]

# Recalculate the 'Age' field
df_cleaned['Age'] = datetime.now().year - df_cleaned['Birth Date'].dt.year

# Adjust age if the birthday hasn't occurred yet this year
df_cleaned['Age'] = df_cleaned['Age'] - ((df_cleaned['Birth Date'].dt.month > datetime.now().month) |
                                         ((df_cleaned['Birth Date'].dt.month == datetime.now().month) &
                                          (df_cleaned['Birth Date'].dt.day > datetime.now().day)))

# Display cleaned data
df_cleaned[['Birth Date', 'Age']].head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Encoding categorical columns
categorical_columns = ['Education Level', 'Occupation', 'Gender', 'Marital Status', 'Home Owner Status', 'City', 'State']
encoder = LabelEncoder()

# Apply encoding to categorical columns
# Use df_cleaned instead of df, since 'City' and 'State' were created in df_cleaned
for col in categorical_columns:
    df_cleaned[col] = encoder.fit_transform(df_cleaned[col].astype(str))

# Check the transformations
df_cleaned.head()
```
